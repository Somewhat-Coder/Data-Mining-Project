```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../Datasets/Electric_Vehicle_Population_Data.csv')
```

```python
df = df.rename(columns={'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'CAFV'})
```

```python
df.columns
```

```python
fig, axes = plt.subplots(2, 2, figsize=(25, 15))

# Boxplot for 'Price'
sns.boxplot(df['Electric Range'], flierprops={"marker": "x"}, ax=axes[0, 0])
# axes[0, 0].set_title('Price')
# axes[0, 0].set_yticklabels(axes[0, 0].get_yticklabels(), fontsize=12)

# Boxplot for 'Engine volume'
sns.boxplot(df['Base MSRP'], flierprops={"marker": "x"}, ax=axes[0, 1])
# axes[0, 1].set_title('Engine Volume')
# axes[0, 1].set_yticklabels(axes[0, 1].get_yticklabels(), fontsize=12)

# Boxplot for 'Airbags'
sns.boxplot(df['2020 Census Tract'], flierprops={"marker": "x"}, ax=axes[1, 0])
# axes[0, 2].set_title('Airbags')
# axes[0, 2].set_yticklabels(axes[0, 2].get_yticklabels(), fontsize=12) 

# Boxplot for 'Prod. year'
sns.boxplot(df['Model Year'], flierprops={"marker": "x"}, ax=axes[1, 1])
# axes[1, 0].set_title('Prod. year')
# axes[1, 0].set_yticklabels(axes[1, 0].get_yticklabels(), fontsize=12)


plt.tight_layout(pad=3)

plt.show()
```

```python
cafv_counts = pd.DataFrame(df["CAFV"].value_counts())
```

```python
cafv_counts
```

```python
plt.figure(figsize=(8, 4)) 
ax = sns.barplot(y='count', data=cafv_counts, palette='viridis', hue='CAFV')
plt.title('CAFV Barplot')

handles, labels = ax.get_legend_handles_labels()
labels = ['Eligibility unknown', 'Eligible', "Not Eligible"]
ax.legend(handles, labels, title='CAFV')

plt.show()
```
