```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../Datasets/nyserda-electric-vehicle-drive-clean-rebate-data-beginning-2017-1.csv')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

sns.boxplot(df['Annual GHG Emissions Reductions (MT CO2e)'], flierprops={"marker": "x"}, ax=axes[0])

sns.boxplot(df['Annual Petroleum Reductions (gallons)'], flierprops={"marker": "x"}, ax=axes[1])

sns.boxplot(df['Rebate Amount (USD)'], flierprops={"marker": "x"}, ax=axes[2])

plt.tight_layout(pad=3)

plt.show()
```

```python
ev_types = pd.DataFrame(df['EV Type'].value_counts())

plt.figure(figsize=(8, 4)) 
ax = sns.barplot(y='count', data=ev_types, palette='viridis', hue='EV Type')
plt.title('EV Type Barplot')

plt.show()
```

```python
ev_types = pd.DataFrame(df['Annual GHG Emissions Reductions (MT CO2e)'].value_counts())

plt.figure(figsize=(8, 4)) 
ax = sns.barplot(y='count', data=ev_types, palette='viridis', hue='Annual GHG Emissions Reductions (MT CO2e)')
plt.title('EV Type Barplot')

plt.show()
```

```python
transaction_types = pd.DataFrame(df['Transaction Type'].value_counts())

plt.figure(figsize=(8, 4)) 
ax = sns.barplot(y='count', data=transaction_types, palette='viridis', hue='Transaction Type')
plt.title('Transaction type Barplot')

plt.show()
```
